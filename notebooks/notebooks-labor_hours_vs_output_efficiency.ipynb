# Labor Hours vs Output Efficiency Analysis

## Objective
Analyze the relationship between labor hours worked and units produced to determine whether increases in labor hours consistently result in proportional output gains and to identify potential inefficiencies or diminishing returns.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Employee_ID": range(1, 11),
    "Hours_Worked": [40, 38, 45, 50, 42, 36, 48, 44, 39, 41],
    "Units_Produced": [400, 370, 420, 460, 410, 350, 445, 430, 390, 405]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.scatter(df["Hours_Worked"], df["Units_Produced"])
plt.xlabel("Hours Worked")
plt.ylabel("Units Produced")
plt.title("Labor Hours vs Output")
plt.show()
```

```python
correlation = df["Hours_Worked"].corr(df["Units_Produced"])
correlation
```

### Key Insight
There is a very strong positive correlation (r â‰ˆ 0.97) between hours worked and units produced, indicating that increases in labor hours are strongly associated with higher output. However, correlation alone does not confirm proportional efficiency gains, highlighting the importance of combining this analysis with productivity-per-hour metrics.
