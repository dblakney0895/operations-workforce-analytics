# Workforce Productivity Analysis

## Objective
Analyze workforce productivity metrics to identify trends, inefficiencies, and opportunities for operational improvement.

This analysis focuses on understanding how labor hours, output, and performance indicators interact to support better staffing and decision-making.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Employee_ID": range(1, 11),
    "Hours_Worked": [40, 38, 45, 50, 42, 36, 48, 44, 39, 41],
    "Units_Produced": [400, 370, 420, 460, 410, 350, 445, 430, 390, 405]
}

df = pd.DataFrame(data)
df
```

```python
df["Productivity"] = df["Units_Produced"] / df["Hours_Worked"]

plt.figure()
plt.bar(df["Employee_ID"], df["Productivity"])
plt.xlabel("Employee ID")
plt.ylabel("Productivity (Units per Hour)")
plt.title("Employee Productivity Comparison")
plt.show()
```
